```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

### https://financeone.com.br/moedas/dolar-hoje/

#### Vamos coletar a página

```python
html=requests.get("https://financeone.com.br/moedas/dolar-hoje/").content
soup=BeautifulSoup(html,'html5lib')
```

```python
linhas=soup.find_all('tr')
linhas
```

```python
df = pd.DataFrame(data=linhas)

df
```
